```python
def zip_scraper(zzzzz):
    '''
    Goal: Obtain all voter percentage of a candidate per locality

    Arguments: Website url (string)

    Returns: Election information per locality from websites (list of lists)
    '''
    url = "https://www.zip-codes.com/zip-code/" + str(zzzzz) + "/zip-code-" + str(zzzzz) + ".asp"
    
    ## Connect to url
    page = requests.get(url)

    ## Parse website
    soup = BeautifulSoup(page.content, 'html.parser')
    
 
    
    ## Get county information
    zip_content = []
    for tag in soup.find_all("a"):
        if (r"/county/" in str(tag)) and ("title" in str(tag)):
            start = str.find(str(tag), ">") + 1
            end = str.find(str(tag), "</a>")
            county = (str(tag)[start:end])
            content = [zzzzz, county]
            zip_content.append(content)
    
    return zip_content
```

```python
va_list = list(va['zip'])

va_list
```

```python
def link_scrape(zip_list, sleep=3):
    '''
    Goal: Apply vde_scraper function to all urls

    Arguments: Set or List containing website urls

    Returns: Data frame containing each locality page's voter preferences
    '''
    ## Create empty list
    scraped_data = []

    ## Loop through each url
    for zzzzz in zip_list:
        ## Apply function...
        scrape = zip_scraper(zzzzz)
        ## If function does not return None...
        if scrape is not None:
            ## Add to empty list
            scraped_data.extend(zip_scraper(zzzzz))
        ## Sleep
        time.sleep(random.uniform(0,sleep))
    ## Convert list of lists to data frame
    df = pd.DataFrame(scraped_data)

    return df

## Store data scraper
va_county = link_scrape(va_list)
```

```python
va_county = va_county.rename(columns={0: "zip", 1: "scraped"})
```

```python
va_county['stcd'] = [x[-2:] for x in va_county['scraped']]
```

```python
va_county['county'] = [x[0:-4] for x in va_county['scraped']]
```
